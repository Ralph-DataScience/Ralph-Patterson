```python
# prompt: function that computes the average of a list of numbers
# implement the unction using a splitted list from the input of user
# call the main function

def average(num_list):
  if not num_list:
    return 0
  return sum(num_list) / len(num_list)

def main():
  numbers_str = input("Enter a list of numbers separated by spaces: ")
  num_list = [int(x) for x in numbers_str.split()]
  avg = average(num_list)
  print("The average is:", avg)

if __name__ == "__main__":
  main()
```
